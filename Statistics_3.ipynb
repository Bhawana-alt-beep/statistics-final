#20 What is an ANOVA test, and what are its assumptions

Assumptions of ANOVA:
Normality: Each sample is drawn from a normally distributed population.
Homogeneity of Variances: The variances among the groups are equal .
Independence: The observations are independent of each other

#3 Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

```python
import numpy as np

def bernoulli_pdf(p, x):
    """
    Calculate the probability distribution function (PDF) of a Bernoulli distribution.

    Parameters:
    p (float): Probability of success (0 <= p <= 1)
    x (int): Outcome (0 or 1)

    Returns:
    float: Probability of the outcome x
    """
    if x not in [0, 1]:
        raise ValueError("Outcome x must be either 0 or 1.")

    return p if x == 1 else 1 - p

# Example usage
p = 0.5  # Probability of success
x = 1    # Outcome

pdf_value = bernoulli_pdf(p, x)
print(f"The PDF value for a Bernoulli distribution with p={p} and outcome x={x} is {pdf_value}.")
```

#9 Write a Python function to calculate Z-scores from a dataset and plot them

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Calculate Z-scores for a dataset.

    Parameters:
    data (list or np.array): The dataset for which to calculate Z-scores.

    Returns:
    np.array: The Z-scores of the dataset.
    """
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev
    return z_scores

# Example dataset
data = [10, 12, 15, 18, 20, 22, 25, 30, 35, 40]

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Plot Z-scores
plt.plot(z_scores, marker='o', linestyle='-', color='b')
plt.title('Z-scores of the Dataset')
plt.xlabel('Data Index')
plt.ylabel('Z-score')
plt.grid(True)
plt.show()

# Print Z-scores
print("Z-scores of the dataset:", z_scores)
```

#10  Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to demonstrate the Central Limit Theorem (CLT)
def demonstrate_clt(distribution, sample_size, num_samples):
    sample_means = []

    for _ in range(num_samples):
        sample = np.random.choice(distribution, sample_size)
        sample_means.append(np.mean(sample))

    return sample_means

# Generate a non-normal distribution (e.g., exponential distribution)
non_normal_distribution = np.random.exponential(scale=2, size=10000)

# Parameters for the CLT demonstration
sample_size = 30
num_samples = 1000

# Get the sample means
sample_means = demonstrate_clt(non_normal_distribution, sample_size, num_samples)

# Plot the original non-normal distribution
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(non_normal_distribution, bins=50, alpha=0.75, color='blue', edgecolor='black')
plt.title('Original Non-Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Plot the distribution of the sample means
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=50, alpha=0.75, color='green', edgecolor='black')
plt.title('Distribution of Sample Means (CLT)')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

#12 Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1)

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_standard_normal_distribution():
    """
    Calculate and plot the standard normal distribution (mean = 0, std = 1).
    """
    # Generate data for the standard normal distribution
    mean = 0
    std_dev = 1
    data = np.random.normal(mean, std_dev, 1000)

    # Plot the histogram of the standard normal distribution
    plt.hist(data, bins=30, density=True, alpha=0.75, edgecolor='black', label='Histogram')

    # Plot the probability density function (PDF) of the standard normal distribution
    x = np.linspace(-4, 4, 1000)
    pdf = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)
    plt.plot(x, pdf, 'r-', lw=2, label='PDF')

    # Add labels and title
    plt.title('Standard Normal Distribution (mean = 0, std = 1)')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Call the function to plot the standard normal distribution
plot_standard_normal_distribution()
```

#14 Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data_point, mean, std_dev):
    """
    Calculate the Z-score for a given data point.

    Parameters:
    data_point (float): The data point for which to calculate the Z-score.
    mean (float): The mean of the dataset.
    std_dev (float): The standard deviation of the dataset.

    Returns:
    float: The Z-score of the data point.
    """
    z_score = (data_point - mean) / std_dev
    return z_score

# Example dataset
data = [10, 12, 15, 18, 20, 22, 25, 30, 35, 40]

# Calculate mean and standard deviation of the dataset
mean = np.mean(data)
std_dev = np.std(data)

# Data point for which to calculate the Z-score
data_point = 25

# Calculate Z-score
z_score = calculate_z_score(data_point, mean, std_dev)

# Plot the standard normal distribution
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

plt.plot(x, y, label='Standard Normal Distribution')
plt.axvline(z_score, color='r', linestyle='--', label=f'Z-score: {z_score:.2f}')
plt.title('Standard Normal Distribution and Z-score')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()

# Print Z-score
print(f"The Z-score for the data point {data_point} is {z_score:.2f}.")
```

#15 Implement hypothesis testing using Z-statistics for a sample datasetE

```python
import numpy as np
from scipy import stats

# Sample dataset
data = [10, 12, 15, 18, 20, 22, 25, 30, 35, 40]

# Calculate the sample mean and standard deviation
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Use ddof=1 for sample standard deviation

# Population mean (hypothesized value)
population_mean = 25

# Sample size
n = len(data)

# Calculate the standard error of the mean
standard_error = sample_std / np.sqrt(n)

# Calculate the Z-score
z_score = (sample_mean - population_mean) / standard_error

# Calculate the p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Print the results
print(f"Sample Mean: {sample_mean}")
print(f"Sample Standard Deviation: {sample_std}")
print(f"Standard Error: {standard_error}")
print(f"Z-score: {z_score}")
print(f"P-value: {p_value}")

# Hypothesis testing result
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

#1 Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results

#4  Perform a two-tailed Z-test using Python and visualize the decision region on a plot

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def one_sample_z_test(sample_data, population_mean, population_std):
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test
    return z_score, p_value

def visualize_decision_region(z_score, alpha=0.05):
    critical_value = stats.norm.ppf(1 - alpha/2)
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x)

    plt.plot(x, y, label='Standard Normal Distribution')
    plt.axvline(x=-critical_value, color='r', linestyle='--', label=f'Critical Value (-{critical_value:.2f})')
    plt.axvline(x=critical_value, color='r', linestyle='--', label=f'Critical Value ({critical_value:.2f})')
    plt.axvline(x=z_score, color='g', linestyle='-', label=f'Z-score ({z_score:.2f})')

    plt.fill_between(x, y, where=(x <= -critical_value) | (x >= critical_value), color='red', alpha=0.3)
    plt.fill_between(x, y, where=(x > -critical_value) & (x < critical_value), color='green', alpha=0.3)

    plt.legend()
    plt.title('Decision Region for Two-Tailed Z-Test')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.show()

# Simulate random data
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=30)
population_mean = 55
population_std = 10

# Perform one-sample Z-test
z_score, p_value = one_sample_z_test(sample_data, population_mean, population_std)
print(f"One-sample Z-test:\nZ-score: {z_score:.2f}\nP-value: {p_value:.4f}")

# Visualize decision region for two-tailed Z-test
visualize_decision_region(z_score)
```

#5 Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_type_errors(alpha=0.05, beta=0.2):
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x)

    critical_value = stats.norm.ppf(1 - alpha/2)

    plt.plot(x, y, label='Standard Normal Distribution')

    # Type 1 error region
    plt.fill_between(x, y, where=(x <= -critical_value) | (x >= critical_value), color='red', alpha=0.3, label='Type 1 Error Region')

    # Type 2 error region
    z_beta = stats.norm.ppf(beta)
    plt.fill_between(x, y, where=(x > -z_beta) & (x < z_beta), color='blue', alpha=0.3, label='Type 2 Error Region')

    plt.legend()
    plt.title('Type 1 and Type 2 Errors in Hypothesis Testing')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.show()

# Visualize Type 1 and Type 2 errors
visualize_type_errors()
```

#9  Write a Python function to calculate the confidence interval for a sample mean and explain its significance

```python
import numpy as np
import scipy.stats as stats

def confidence_interval(sample_data, confidence_level=0.95):
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation
    alpha = 1 - confidence_level
    t_critical = stats.t.ppf(1 - alpha/2, df=sample_size-1)  # T-critical value for two-tailed test
    margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))
    confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)
    return confidence_interval
# Simulate random data
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=30)

# Calculate confidence interval
ci = confidence_interval(sample_data)
print(f"Confidence Interval: {ci}")
```

#11 Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

```python
import numpy as np

def bayesian_inference(prior, likelihood, evidence):
    posterior = (likelihood * prior) / evidence
    return posterior

# Example: Bayesian inference for a medical test
# Prior probability (P(Disease))
prior_disease = 0.01

# Likelihood (P(Test Positive | Disease))
likelihood_test_positive_given_disease = 0.99

# Evidence (P(Test Positive))
# P(Test Positive) = P(Test Positive | Disease) * P(Disease) + P(Test Positive | No Disease) * P(No Disease)
prob_test_positive_given_no_disease = 0.05
prob_no_disease = 1 - prior_disease
evidence_test_positive = (likelihood_test_positive_given_disease * prior_disease) + (prob_test_positive_given_no_disease * prob_no_disease)

# Posterior probability (P(Disease | Test Positive))
posterior_disease_given_test_positive = bayesian_inference(prior_disease, likelihood_test_positive_given_disease, evidence_test_positive)

print(f"Posterior Probability of Disease given Test Positive: {posterior_disease_given_test_positive:.4f}")
```

#13 Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data

```python
import numpy as np

def calculate_expected_frequencies(observed):
    """
    Calculate the expected frequencies for a Chi-square test based on observed data.

    Parameters:
    observed (2D list): A 2D list where each sublist represents the observed frequencies in a category.

    Returns:
    np.array: A 2D array of expected frequencies.
    """
    observed = np.array(observed)
    row_totals = observed.sum(axis=1)
    col_totals = observed.sum(axis=0)
    total = observed.sum()

    expected = np.outer(row_totals, col_totals) / total
    return expected

# Example usage
observed_data = [
    [10, 20, 30],
    [6, 9, 12]
]

expected_frequencies = calculate_expected_frequencies(observed_data)
print("Expected Frequencies:")
print(expected_frequencies)
```
